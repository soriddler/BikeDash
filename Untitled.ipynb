# JSON as data source

```python
def flatten_json(y):
    out = {}
    def flatten(x, name =''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            for i, a in enumerate(x):                
                flatten(a, name + str(i) + '_')
        else:
            out[name[:-1]] = x
    flatten(y)
    return out
```

```python
import json
import pandas as pd

f = open('GarminDump/activities-1-964.json')
data = json.load(f)
flattened_data = []
for activity in data:
    activity["path"] = f'GarminDump/activity_{activity["activityId"]}.gpx'
    flattened_data.append(flatten_json(activity))
    
# Serializing json 
json_object = json.dumps(flattened_data, indent = 4)
  
# Writing to sample.json
with open("GarminDump.json", "w") as outfile:
    outfile.write(json_object)
```
